```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# Load data
file_path = "/Users/shaonan.feng/Documents/Self-learning/DA Projects/MA_Analysis_2025/Deals_Results_report.xlsx"
sheet_name = "Transactions_Results"

df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=3)
# Display first few rows
df.head(5)
```

```python
# inspect and rename columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^0-9a-zA-Z_]', '', regex=True)
df.head(5)
```

```python
# check for missing values
df = df.replace("-", np.nan)
df.isna().mean().sort_values(ascending=False).head(10)
```

```python
# drop fully empty columns
df.dropna(axis=1, how='all', inplace=True)
```

```python
# Convert data types
df['announcement_date'] = pd.to_datetime(df['announcement_date'], errors='coerce')
df['year'] = df['announcement_date'].dt.year
df['month'] = df['announcement_date'].dt.month
df['quarter'] = df['announcement_date'].dt.quarter
```

```python
# Filter dataset for 2025 Q1-Q3 deals
df = df[(df['year'] == 2025) & (df['quarter'].isin([1, 2, 3]))]
```

```python
df['sector_targetissuer'] = df['sector_targetissuer'].str.title().str.strip()
df['factset_sector_buyerinvestor'] = df['factset_sector_buyerinvestor'].str.title().str.strip()
```

```python
# sanity check
print(df.shape)
df['year'].value_counts().sort_index()
df.describe(include='all')
```

```python
# Save cleaned data
output_path = "/Users/shaonan.feng/Documents/Self-learning/DA Projects/MA_Analysis_2025/ma_germany_2025_cleaned.csv"
df.to_csv(output_path, index=False)
print("Cleaned data saved to:", output_path)
```

```python
# Compute key indicators
import pandas as pd
df = pd.read_csv("/Users/shaonan.feng/Documents/Self-learning/DA Projects/MA_Analysis_2025/ma_germany_2025_cleaned.csv", parse_dates=['announcement_date'])

total_deals = len(df)
total_deal_size_mm_usd = df['deal_size_mm_usd'].sum(min_count=1)  # may be NaN if many missing
cross_border_pct = ( (df['country_buyerinvestor'].str.lower() != df['country_targetissuer'].str.lower()) ).mean() * 100

kpi_df = pd.DataFrame({
    'key indicator': ['total_deals', 'total_deal_size_mm_usd', 'cross_border_pct'],
    'value': [total_deals, total_deal_size_mm_usd, cross_border_pct]
})
kpi_df
```

## Methods & Data

**Source:** FactSet Transactions_Results export (sheet: Transactions_Results).  
**Data cleaning highlights:**
- Skipped first 3 export rows and used the 4th as header.  
- Replaced missing values '-' with `NaN`.  
- Filtered to announcement_date between 2025-01-01 and 2025-09-30.  
- Structured transactions to quarters.
